Let's calculate and compare minimum dsistances between points in 1D and 2D.

```
oneD_distances = np.array([np.linalg.norm(point - oneD[:], axis=-1) for point in oneD])
print(oneD_distances.shape)
print(oneD_distances[:5,:5])
```

We will use `triu_indices` function (with an offset 1 from the main diagonal) to extract upper triangle of the distance matrix and take minimum of these

```
min1D_distance = np.min(oneD_distances[np.triu_indices(len(oneD), k = 1)])
print(min1D_distance)
```

```
##### YOUR CODE STARTS #####
twoD_distances = np.array([np.linalg.norm(point - twoD[:], axis=-1) for point in twoD])
min2D_distance = np.min(twoD_distances[np.triu_indices(len(twoD), k = 1)])
##### YOUR CODE ENDS ##### (please do not delete this line)
print(min2D_distance)
```

It has increased by at least one order of magnitude!

Make a function that samples a given number of points with a fixed number of dimensions from the random distribution and then calculates the minimum distance between these points (copy and paste from already created code).

```
##### YOUR CODE STARTS #####
def compute_min_distance(n_points, n_dimensions):
  points = np.random.rand(n_points, n_dimensions)
  distances = np.array([np.linalg.norm(point - points[:], axis=-1) for point in points])
  min_distance = np.min(distances[np.triu_indices(len(distances), k = 1)])
  return(min_distance)
##### YOUR CODE ENDS ##### (please do not delete this line)
```

What about minimum smallest distance between points in 3D space?

Transpose the matrix of original coordinates

```
Z = np.matrix(Z)
Z_t = Z.T
print(Z)
print(Z_t)
```

Compute covariance matrix

```
##### YOUR CODE STARTS #####
S = np.dot(Z_t, Z)/4
##### YOUR CODE ENDS ##### (please do not delete this line)
print(S)
```

Perform eigendecomposition using `linalg.eig` function from `numpy`

```
eigenvalues, eigenvectors = np.linalg.eig(S)
print(eigenvalues)
print(-1*eigenvectors)
```

Transform transposed matrix of original coordinates into the space of principle components.

Answer (g): 


1.   By compressing the number of features, PCA decreases both the training time and reduces the overfitting problem, since we are eliminating the not-so-useful features from our model. Also, since PCA gets rid of the correlation between the features by transforming them into independent principal components, it is easier for predictive models to understand the effects of the principal components on the target variable, thus improving the model accuracy.

2.   It makes model interpretability very difficult since it destroys the original features by mixing them together based on the axes that they vary the most.

3.   PCA comes handy when we have an enormous number of features. Since, in general, not all of those features are equally important it is good to reduce the dimensions of our training data by eliminating some of the features. 

  If the feature number is low and all features are more or less important, PCA will not make much improvements for our model accuracy. In addition to this, if we care about model interpratibility, we shouldn't use PCA in any case, whether our data is high dimensional or not.

```
##### YOUR CODE STARTS #####
def compute_inertia(clustering, cluster_data):
  sse = np.zeros(len(clustering.cluster_centers_))
  for (index, centroid) in enumerate(clustering.cluster_centers_):
    sse[index] = np.sum(dist(cluster_data[index == clustering.labels_], centroid) ** 2)
  return(np.sum(sse))
##### YOUR CODE ENDS ##### (please do not delete this line)
```
